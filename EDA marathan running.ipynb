### Read and understand the dataset

```python
import pandas as pd
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & 
   (df['Year of event'] == 2020) &
   (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')
]
```

```python
df = df[(df['Event distance/length'].isin(['50km', '50mi'])) & 
   (df['Year of event'] == 2020) &
   (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')
   ]
```

```python
df.shape
```

```python
df['Event name'] = df['Event name'].str.split('(').str.get(0)
```

```python
df
```

```python
df['Athlete age']  = 2020 - df['Athlete year of birth']
```

```python
df['Athlete performance'] = df['Athlete performance'].str.split(' ').str.get(0)
```

```python
df = df.drop(columns=['Athlete club','Athlete country', 'Athlete age category', 'Athlete year of birth' ])
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['Athlete age'] = df['Athlete age'].astype(int)
df['Athlete average speed'] = df['Athlete average speed'].astype(float)
```

```python
df.dtypes
```

```python
df
```

```python
import seaborn as sns
sns.countplot(data=df, x='Event distance/length', hue= 'Athlete gender')
```

```python
sns.violinplot(data=df, x = 'Event distance/length', y ='Athlete average speed' , hue= 'Athlete gender')
```

```python
sns.lineplot(data=df, x = 'Athlete age', y = 'Athlete average speed', hue = 'Athlete gender')
```

```python
df.groupby(['Event distance/length', 'Athlete gender'])['Athlete average speed'].mean()
```

```python
df['Race month'] = df['Event dates'].str.split('.').str.get(1).astype(int)
df
```

```python
df['Race season'] = df['Race month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df
```

```python
sns.countplot(df, x= 'Race season')
```

```python
df.groupby(['Event name'])['Event number of finishers'].count().sort_values(ascending=False)
```

```python
data = df.groupby(['Event distance/length'])['Event number of finishers'].count().sort_values(ascending=False)
sns.barplot(data =  data, palette= 'mako')
```
